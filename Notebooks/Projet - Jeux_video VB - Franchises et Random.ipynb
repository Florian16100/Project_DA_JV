```python
# Import de vgsales et de la liste de franchises populaires dans des DataFrames

vgsales = pd.read_csv('vgsales_cleaned.csv')
franchises = pd.read_csv('popular_vg_franchises.csv', encoding = 'latin-1')

display(vgsales.head())
display(franchises.head())
```

```python
# Définition d'une fonction utilisant du text mining pour repérer la license à laquelle appartient chaque nom de jeu
# Application de la fonction sur le DataFrame vgsales

def franchise_finder(vgname, vgfranchises):
    for franchise in vgfranchises:
        match = re.search(franchise, vgname)
        if match:
            return match.group()
    return 'Undefined Franchise'

vgsales['Franchise'] = vgsales['Name'].apply(franchise_finder, vgfranchises=franchises['Franchise name'])
```

```python
# Création d'une colonne contenant des valeurs aléatoires, à utiliser pour faire des tests lors de la phase de machine learning

from numpy import random
vgsales["Random_Values"] = random.rand(len(vgsales.index))
```
