```python
# Création du DataFrame

vgsales = pd.read_csv('vgsales-12-4-2019.csv')
vgsales.head()
```

```python
# Informations sur le DataFrame

vgsales.info()
```

```python
# Création d'une variable additionnelle prenant par défaut la valeur de Global Sales. Si manquante, la valeur de Total Shipped.
# Si toujours manquante dans les deux variable, prend la valeur np.nan

def estimated_sales_values (row):
    if not pd.isna(row['Total_Shipped']):
        return row['Total_Shipped']
    elif not pd.isna(row['Global_Sales']):
        return row['Global_Sales']
    else:
        return np.nan
    
vgsales['Estimated_Sales'] = vgsales.apply(estimated_sales_values, axis = 1)

vgsales.head()
```

```python
# Certaines colonnes ne nous seront d'aucune utilité pour l'analyse. Création d'un nouveau DataFrame ne contenant que les données utiles pour l'analyse.
# État des valeurs manquantes.

vgsales_new = vgsales[['Rank', 'Name', 'basename', 'Platform', 'Publisher', 'Developer', 'Year', 'Estimated_Sales']].copy()

display(vgsales_new.head())
print("\nRapport de valeurs manquantes:\n")
print(vgsales_new.isna().sum())
print("\nNombre total d'entrées:", len(vgsales_new))
```

```python
# Remplacement des valeurs 'Unknown' de la variable Publisher et Developer par np.nan

vgsales_new['Publisher'] = vgsales_new['Publisher'].apply(lambda x: np.nan if x == 'Unknown' else x)
vgsales_new['Developer'] = vgsales_new['Developer'].apply(lambda x: np.nan if x == 'Unknown' else x)

vgsales_new.isna().sum()
```

```python
# Navigation dans les valeurs manquantes pour définir une stratégie de cleaning

vgsales_new[(vgsales_new['Year'].isna() == True)].head(30)
```

```python
# Création de deux fonctions pour nettoyer les données des variables Developer et Publisher
# La stratégie est de récupérer le nom de chaque jeu comportant des valeurs manquantes pour les variables Developer et Publisher
# On crée une Series pandas à partir de chaque nom récupéré. On en définit ensuite la mode. 
# On remplace finalement les valeurs manquante par la mode définit par la fonction. 

def developer_cleaning (row):
    if pd.isna(row['Developer']):
        Name = row['Name']
        Developer_Series = vgsales_new[(vgsales_new['Name'] == Name)]['Developer']
        Developer_Mode = Developer_Series.mode()
        if not Developer_Mode.empty:
            return Developer_Mode.iloc[0]
        else:
            return np.nan
    else:
        return row['Developer']
    
    
def publisher_cleaning (row):
    if pd.isna(row['Publisher']):
        Name = row['Name']
        Publisher_Series = vgsales_new[(vgsales_new['Name'] == Name)]['Publisher']
        Publisher_Mode = Publisher_Series.mode()
        if not Publisher_Mode.empty:
            return Publisher_Mode.iloc[0]
        else:
            return np.nan
    else:
        return row['Publisher']
    

vgsales_new['Developer'] = vgsales_new.apply(developer_cleaning, axis = 1)
vgsales_new['Publisher'] = vgsales_new.apply(publisher_cleaning, axis = 1)

vgsales_new.isna().sum()

# Correction de près de 1'000 valeurs manquantes par variable

# Autre manière de faire d'apère Chat GPT, plus synthétique:
# vgsales_new['Developer'] = vgsales_new['Developer'].fillna(vgsales_new.groupby('Name')['Developer'].transform(lambda x: x.mode()[0]))
# vgsales_new['Publisher'] = vgsales_new['Publisher'].fillna(vgsales_new.groupby('Name')['Publisher'].transform(lambda x: x.mode()[0]))
```

```python
# Vérification du nombre de lignes comportant 3 valeurs manquantes ou plus

display(vgsales_new[vgsales_new.isna().sum(axis = 1) >= 3].head())
print(len(vgsales_new[vgsales_new.isna().sum(axis = 1) >= 3]))

# Plus 1400 lignes sont concernées. Elles sont trop de valeur manquantes, nous pouvont les supprimer.

vgsales_new = vgsales_new.dropna(thresh = vgsales_new.shape[1] - 2)

# 0 lignes comportant 3 valeurs manquantes ou plus
```

```python
# Nouvelle somme des valeurs manquantes

vgsales_new.isna().sum()
```

```python
# Suppression des lignes n'ayant pas de valeur dans la colonne Estimated Sales.
# Comme Estimated Sales sera notre variable cible, garder des valeurs manquantes dans cette colonne compliquerait inutilement le travail du modèle.

vgsales_new = vgsales_new.dropna(subset = ['Estimated_Sales'])
```

```python
# Nouvelle somme des valeurs manquantes

print(vgsales_new.isna().sum(), "\n")
print(len(vgsales_new))

# Il ne reste que quelques valeurs manquantes dans les colonnes Developer, Publisher et Year.
# Nous ne les supprimons pas, représentant de l'information utile lors de la modélisation.
```

```python
# Instanciation d'un nouveau DataFrame

vgsales_cleaned = vgsales_new
```
